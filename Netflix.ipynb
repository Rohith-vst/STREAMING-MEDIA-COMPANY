```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('netflix.csv')
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop('description',axis=1,inplace= True)
df.head(10)
```

```python
df.isnull().sum()/len(df)*100
```

```python
df['country'].nunique()
```

```python
df=df[df['country'].notna()]
```

# unnesting of dataframe

```python
df1 = df.copy(deep=True)
```

```python
split_country=df1['country'].str.split(',')
split_country
```

```python
df1["splitted_country"]= split_country
df1.drop("country",axis =1,inplace=True)
```

```python
df1=df1.explode('splitted_country')

df1.head()
```

```python
df1['genre']=df1['listed_in'].str.split(',')
df1=df1.explode('genre')
df1.drop('listed_in',axis = 1 , inplace = True)
df1.head()
```

```python
df1['Cast']=df1['cast'].str.split(',')
df1=df1.explode('Cast')
df1.drop('cast',axis = 1 , inplace = True)
df1.head()
```

# removing null values

```python
df1.isnull().sum()/len(df)*100
```

```python
df1=df1[df1['Cast'].notna()]
df1=df1[df1['date_added'].notna()]
df1=df1[df1['duration'].notna()]
df1=df1[df1['rating'].notna()]
```

```python
df1.isnull().sum()/len(df)*100
```

```python
#filling mode of directors as  per country
```

```python
df1
```

```python
f = lambda x: x.mode().iat[0] if x.notna().any() else np.nan
df1['director'] = df1['director'].fillna(df1.groupby('splitted_country')['director'].transform(f))
df1.head()
```

```python
df1.info()
```

# changing data types

```python
#changing format of 'duration' column
```

```python
df1['duration_new']=df1['duration'].apply(lambda x : str(x).split(' ')[0])
```

```python
df1.drop('duration',axis = 1,inplace = True)
```

```python
#changing the 'object' datatype to 'date time' format
```

```python
df1["date_added"] = pd.to_datetime(df1["date_added"])
```

```python
#changing data type of 'duartion_new' column from object to int32
```

```python
df1 ['duration_new'] = df1 ['duration_new'].astype(str).astype(int) 
```

```python
df1.head(20)
```

```python
df2=df1.pivot_table(columns='type',index = 'release_year',values = 'duration_new',aggfunc='mean').fillna(0)
df2['TV Show']=np.round(df2['TV Show'])
df2.head(20)
```

```python
df3=df1.pivot_table(columns='splitted_country',index = 'genre',values = 'duration_new',aggfunc='mean').fillna(0)
df3.head(20)
```

```python
df3.info()
```
